```python
import pandas as pd
import numpy as np

# Load the embeddings
# Convert strings back to numpy arrays
def string_to_array(vector_str):
    return np.array([float(x) for x in vector_str.split(',')])

df_loaded = pd.read_csv('processed_embeddings_1.csv')
df_loaded
```

```python
df_loaded['abstract_vector'] = df_loaded['abstract_vector'].apply(string_to_array)
df_loaded
```

```python
# Verify the conversion worked
print("First vector type:", type(df_loaded['abstract_vector'].iloc[0]))
print("First vector shape:", df_loaded['abstract_vector'].iloc[0].shape)
```

```python
# Check length of abstract vectors
print("Length of first abstract vector:", len(df_loaded['abstract_vector'].iloc[0]))

# Verify all vectors have same length
vector_lengths = df_loaded['abstract_vector'].apply(len)
print("\nAll vectors have same length:", all(vector_lengths == vector_lengths.iloc[0]))
print("Vector lengths:", vector_lengths.unique())
```
